# Notebook Imports

```python
from sklearn.manifold import MDS
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import manifold, datasets
```

# Generate Data points

```python
ax = plt.axes(projection='3d')
n_pts = 1000
X, color = datasets.make_s_curve(n_pts, random_state=0)
ax.scatter3D(X[:, 0], X[:, 1], c=color)
plt.show()
```

## Project the Data points of the Swiss roll into a two-dimensional feature space using MDS

```python
n_components = 2
mds = MDS(2, random_state=0)
X_2d = mds.fit_transform(X)

# Display the obtained projection in the two-dimensional feature space
plt.scatter(X_2d[:,0], X_2d[:,1], c=color)
plt.show()
```
