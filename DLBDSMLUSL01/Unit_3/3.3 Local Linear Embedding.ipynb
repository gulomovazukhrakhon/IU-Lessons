# Notbook Imports

```python
import matplotlib.pyplot as plt 
from sklearn.manifold import LocallyLinearEmbedding
from sklearn import manifold, datasets
```

## Generate Sample Data

```python
n_pts = 1000
X, color = datasets.make_s_curve(n_pts, random_state=0)
```

## Model

```python
embedding = LocallyLinearEmbedding(n_neighbors=12, n_components=2) # project it into a 2-dimensional feature space
X_2d = embedding.fit_transform(X)
```

## Visualization

```python
plt.scatter(X_2d[:,0], X_2d[:,1], c=color)
plt.show()
```
