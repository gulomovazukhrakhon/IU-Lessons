```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
```

## Generate Data

```python
table = {
    'Customer-ID': [1, 2, 3, 4, 5], 
    'Gender': ['M', 'F', 'M',' F', 'F'], 
    'Work-type': [1, 2, 2, 3, 3], 
    'Client-satisfaction': [3, 0, 4, 3, 5], 
    'Number-occupants': [2, 4, 2, 1, 2], 
    'Consumption': [70, 140, 65, 40, 65]
}

TDF = pd.DataFrame(data=table)
TDF
```

## Min-Max Scaling

```python
MMS = MinMaxScaler().fit_transform(TDF[['Consumption']])
TDF['Consumption'] = MMS
 
TDF[['Consumption']]
```

```python
TDF[['Consumption']].describe()
```

## Variance Scaling

```python
TDF = pd.DataFrame(data=table)
ST = StandardScaler().fit_transform(TDF[['Consumption']])
TDF['Consumption'] = ST
 
TDF[['Consumption']]
```

```python
TDF[['Consumption']].describe()
```

## Robust Scaling

```python
TDF = pd.DataFrame(data=table)
RS = RobustScaler().fit_transform(TDF[['Consumption']])
TDF['Consumption'] = RS
TDF
```
